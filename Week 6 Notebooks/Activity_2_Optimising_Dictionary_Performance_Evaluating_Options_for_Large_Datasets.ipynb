## Extension: Optimising Dictionary Performance in a Real-World Scenario

You are working on a data analysis project that involves processing large log files to extract and summarise relevant information. The log files contain millions of entries, each with a timestamp, user ID, and other metadata.

**Task:** Design an efficient data processing pipeline that utilises dictionary-based data structures to store and manipulate the log data. Justify your choices of dictionary implementations and explain how they address the performance requirements of this scenario.

Consider factors such as:
- Handling large datasets with millions of entries
- Efficiently grouping and aggregating data by user, timestamp, or other metadata
- Optimising lookup and update operations for the analysis tasks
- Trade-offs between time complexity, space usage, and ease of implementation

```python
import datetime
import random
from collections import defaultdict

# --- 1. Generating Sample Log Data ---
def generate_log_entries(num_entries):
    """Generates a list of simulated log entries."""
    log_data = []
    for i in range(num_entries):
        timestamp = datetime.datetime.now() - datetime.timedelta(minutes=random.randint(0, 60 * 24 * 7))
        user_id = f"user_{random.randint(1, 1000)}"
        event_type = random.choice(["login", "logout", "purchase", "view_product", "error"])
        metadata = {"ip_address": f"192.168.1.{random.randint(1, 254)}", "session_id": f"sess_{random.randint(1000, 9999)}"}
        log_data.append({
            "timestamp": timestamp.isoformat(),
            "user_id": user_id,
            "event_type": event_type,
            "metadata": metadata
        })
    return log_data

# --- 2. Designing the Data Processing Pipeline ---
def process_log_data(log_entries):
    """Processes log entries to group and aggregate data using dictionaries."""

    # Use a defaultdict for efficient grouping by user_id
    # This avoids checking if a key exists before appending, simplifying logic
    # and slightly improving performance for large datasets.
    user_logs = defaultdict(list)
    user_summary = {}

    print(f"Processing {len(log_entries)} log entries...")

    for entry in log_entries:
        user_id = entry['user_id']
        timestamp = datetime.datetime.fromisoformat(entry['timestamp'])

        # Group all raw log entries for a user
        user_logs[user_id].append(entry)

        # Maintain a summary for quick lookups and aggregations
        if user_id not in user_summary:
            user_summary[user_id] = {
                'total_events': 0,
                'first_activity': timestamp,
                'last_activity': timestamp,
                'event_counts': defaultdict(int)
            }

        # Update summary statistics
        user_summary[user_id]['total_events'] += 1
        user_summary[user_id]['first_activity'] = min(user_summary[user_id]['first_activity'], timestamp)
        user_summary[user_id]['last_activity'] = max(user_summary[user_id]['last_activity'], timestamp)
        user_summary[user_id]['event_counts'][entry['event_type']] += 1

    # Convert defaultdicts to regular dicts for final output if preferred
    for user_id in user_summary:
        user_summary[user_id]['event_counts'] = dict(user_summary[user_id]['event_counts'])

    print("Processing complete.")
    return dict(user_logs), user_summary

# --- 3. Example Usage and Justification ---
if __name__ == "__main__":
    # Simulate a large dataset
    NUM_LOG_ENTRIES = 1_000_000 # Millions of entries
    print(f"Generating {NUM_LOG_ENTRIES} log entries...")
    log_entries_data = generate_log_entries(NUM_LOG_ENTRIES)
    print("Log generation complete.\n")

    grouped_logs, aggregated_summary = process_log_data(log_entries_data)

    print("\n--- Analysis Results ---")
    print(f"Total unique users found: {len(aggregated_summary)}")

    # Example: Look up data for a specific user
    sample_user_id = f"user_{random.randint(1, 1000)}"
    if sample_user_id in aggregated_summary:
        print(f"\nSummary for {sample_user_id}:")
        print(f"  Total events: {aggregated_summary[sample_user_id]['total_events']}")
        print(f"  First activity: {aggregated_summary[sample_user_id]['first_activity']}")
        print(f"  Last activity: {aggregated_summary[sample_user_id]['last_activity']}")
        print(f"  Event counts: {aggregated_summary[sample_user_id]['event_counts']}")
    else:
        print(f"\nNo data found for {sample_user_id}")

    # Justification of Choices:
    # 1. Handling large datasets: Python's built-in dictionaries (dict and defaultdict) are implemented as hash tables.
    #    They provide average O(1) time complexity for insertions, deletions, and lookups.
    #    This makes them highly efficient even with millions of entries, as operations don't scale linearly with dataset size.
    #    For truly massive datasets that exceed memory, external solutions (like databases or distributed systems) would be needed,
    #    but for in-memory processing up to several million entries, Python's dicts are very performant.

    # 2. Efficiently grouping and aggregating data:
    #    - `defaultdict(list)`: This is ideal for grouping. When a new user_id is encountered, it automatically creates an empty list,
    #      allowing direct appending of log entries without explicit `if key not in dict` checks. This simplifies code and improves efficiency.
    #    - `dict` for `user_summary`: A standard dictionary is used to store aggregated statistics per user. The user_id acts as the key,
    #      providing O(1) average-case lookup to update summaries (`total_events`, `first_activity`, etc.).
    #    - `defaultdict(int)` for `event_counts`: Within each user's summary, another `defaultdict(int)` is used to count specific event types,
    #      again leveraging O(1) average-case updates for event type counts.

    # 3. Optimising lookup and update operations:
    #    - All core operations (grouping, updating counts, min/max timestamps) rely on dictionary lookups and assignments, which are O(1) on average.
    #    - This ensures that as the number of log entries grows, the time taken per entry remains constant, leading to efficient processing overall.
    #    - After processing, accessing any user's summary or their grouped logs is also an O(1) operation.

    # 4. Trade-offs:
    #    - Time Complexity: Excellent (average O(1) for key operations).
    #    - Space Usage: The primary trade-off. Storing all log entries and their summaries in memory can consume significant RAM for millions of detailed entries.
    #      For extremely sparse data or if memory is highly constrained, other structures like `shelve` (disk-backed dict) or external databases might be considered.
    #      However, for typical server memory configurations (e.g., 8GB+), processing millions of log summaries is often feasible.
    #    - Ease of Implementation: Very high. Python's dictionaries are intuitive and powerful, allowing for concise and readable code for these tasks.

    # This pipeline effectively uses Python's dictionary implementations to provide a robust and performant solution
    # for processing and analysing large log datasets, balancing efficiency with ease of development.
```
