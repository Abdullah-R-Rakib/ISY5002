## Activity 1: Conceptual Application & Analysis

You've been tasked with analysing sales data for a small retail business. The data is stored in a dictionary, where the keys represent product IDs and the values are dictionaries containing information about each product, such as its name, price, and quantity sold.

```python
sales_data = {
    "P001": {"name": "Widget A", "price": 9.99, "qty_sold": 120},
    "P002": {"name": "Widget B", "price": 14.50, "qty_sold": 85},
    "P003": {"name": "Widget C", "price": 7.25, "qty_sold": 200},
    "P004": {"name": "Widget D", "price": 12.75, "qty_sold": 75}
}
```

1. Explain how you would use dictionary operations and methods to calculate the total revenue generated from the sales data. Describe the step-by-step process, including any potential challenges or considerations.

💡 AI Tip: Ask an AI to suggest alternative approaches to calculating the total revenue using dictionaries. Compare the AI's suggestions to your own process and identify any additional factors you should consider.

2. The business owner wants to identify the best-selling product. Describe how you would use dictionary operations and methods to determine the product with the highest quantity sold, and explain your reasoning.


```python
# Sales data
sales_data = {
    "P001": {"name": "Widget A", "price": 9.99, "qty_sold": 120},
    "P002": {"name": "Widget B", "price": 14.50, "qty_sold": 85},
    "P003": {"name": "Widget C", "price": 7.25, "qty_sold": 200},
    "P004": {"name": "Widget D", "price": 12.75, "qty_sold": 75}
}

# 1. Calculate total revenue
total_revenue = 0

for details in sales_data.values():
    total_revenue += details["price"] * details["qty_sold"]

print("Total Revenue:", total_revenue)


# 2. Find best-selling product
max_qty = 0
best_product = ""

for details in sales_data.values():
    if details["qty_sold"] > max_qty:
        max_qty = details["qty_sold"]
        best_product = details["name"]

print("Best-selling product:", best_product)
```


## Activity 2: Evaluation, Comparison, and Justification

The business owner now wants to expand their product line and is considering adding a new product. They've provided you with the following information about two potential new products:

Product A:
- Price: $19.99
- Estimated monthly sales: 50 units

Product B:
- Price: $14.75
- Estimated monthly sales: 80 units

Using the sales data dictionary from the previous activity, evaluate which of the two new products would be a better addition to the business. Justify your recommendation by comparing the potential revenue and profitability of each new product.

💡 AI Tip: Ask an AI to identify potential factors you should consider when evaluating the suitability of the new products, beyond just revenue and profitability. Critically assess the AI's suggestions and incorporate them into your justification.



```python
# New product information
product_a = {"name": "Product A", "price": 19.99, "estimated_monthly_sales": 50}
product_b = {"name": "Product B", "price": 14.75, "estimated_monthly_sales": 80}

# Calculate potential revenue for Product A
revenue_a = product_a["price"] * product_a["estimated_monthly_sales"]
print(f"Potential monthly revenue for {product_a['name']}: ${revenue_a:.2f}")

# Calculate potential revenue for Product B
revenue_b = product_b["price"] * product_b["estimated_monthly_sales"]
print(f"Potential monthly revenue for {product_b['name']}: ${revenue_b:.2f}")

# Justify the recommendation
if revenue_a > revenue_b:
    recommendation = product_a['name']
    higher_revenue = revenue_a
    lower_revenue = revenue_b
    other_product = product_b['name']
else:
    recommendation = product_b['name']
    higher_revenue = revenue_b
    lower_revenue = revenue_a
    other_product = product_a['name']

print(f"\nRecommendation: {recommendation} would be a better addition.")
print(f"Justification: {recommendation} is projected to generate ${higher_revenue:.2f} in monthly revenue, which is higher than {other_product}'s projected ${lower_revenue:.2f}.")
print("This suggests a greater immediate revenue potential for the business. However, other factors like cost of goods sold, market demand, and strategic fit would also need to be considered for a complete profitability analysis.")
```

## Extension: Combining Concepts

The business owner is now interested in understanding the overall sales performance of their product line. They've asked you to create a summary report that provides the following insights:

1. The total revenue generated from all products.
2. The product with the highest total revenue.
3. The average price of all products.

Using the sales data dictionary, write code to calculate these metrics and provide a brief explanation of your approach, including any relevant dictionary operations or methods you used.

💡 AI Tip: If you used AI to help generate the code or explanations for this task, describe how you verified the accuracy of the AI's output and adapted it to fit the specific requirements of the report.



```python
# 1. Calculate the total revenue from all products
total_revenue_all_products = 0
for product_id, details in sales_data.items():
    total_revenue_all_products += details['price'] * details['qty_sold']

print(f"Total revenue from all products: ${total_revenue_all_products:.2f}")

# 2. Find the product with the highest total revenue
highest_revenue = 0
product_with_highest_revenue = ''

for product_id, details in sales_data.items():
    current_product_revenue = details['price'] * details['qty_sold']
    if current_product_revenue > highest_revenue:
        highest_revenue = current_product_revenue
        product_with_highest_revenue = details['name']

print(f"Product with the highest total revenue: {product_with_highest_revenue} (${highest_revenue:.2f})")

# 3. Calculate the average price of all products
total_price = 0
num_products = 0

for product_id, details in sales_data.items():
    total_price += details['price']
    num_products += 1

average_price = total_price / num_products

print(f"Average price of all products: ${average_price:.2f}")

print("\nApproach Explanation:")
print("To calculate the total revenue, I iterated through the `sales_data` dictionary using `.items()` to access each product's details. For each product, I multiplied its 'price' by its 'qty_sold' and added it to a running total.")
print("To find the product with the highest total revenue, I again iterated through the `sales_data`. In each iteration, I calculated the current product's revenue and compared it with a `highest_revenue` variable, updating both `highest_revenue` and `product_with_highest_revenue` if a new maximum was found.")
print("To calculate the average price, I summed up the 'price' of each product and divided by the total number of products, which I counted during the iteration. The `.items()` method was crucial for accessing both product IDs and their corresponding detail dictionaries efficiently.")
```
