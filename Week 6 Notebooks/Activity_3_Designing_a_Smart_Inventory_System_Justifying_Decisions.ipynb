### Activity 1: Conceptual Application & Analysis

You have been tasked with designing an inventory management system for a small retail business that sells a variety of products. The business owner wants to use a dictionary-powered approach to track inventory levels and generate reports.

Explain the process you would use to design this inventory management system. In your explanation, address the following:

- What are the key considerations you would take into account when choosing a dictionary-based approach?
- What potential challenges or limitations might you encounter when using a dictionary for inventory management, and how would you address them?
- How would you structure the dictionary to efficiently store and retrieve inventory data?

```python
# I would design the inventory system using a dictionary where each product ID is the key, and the value is another dictionary containing product details like name, price, stock quantity, and category. The main reason for choosing a dictionary-based approach is that it allows fast lookup and updates (O(1) on average), which is important for tracking stock levels in real time. Key considerations include ensuring efficient access, avoiding duplicate product IDs, and making the structure scalable as the inventory grows.

#There are some limitations. For example, dictionaries can become memory-heavy when storing large inventories, and they do not handle complex relationships or reporting as efficiently as databases. There is also a risk of data inconsistency if updates are not managed properly. To address this, I would validate inputs carefully, use functions to manage updates, and consider integrating a database if the system grows.

#The dictionary would be structured like this: {product_id: {"name": ..., "price": ..., "stock": ..., "category": ...}}, which allows easy access to any product’s details and quick updates to stock levels when sales or restocking occur.
```

### Activity 2: Evaluation, Comparison, and Justification

The business owner has asked you to compare two different inventory management strategies: just-in-time (JIT) and economic order quantity (EOQ). They want you to evaluate the suitability of each approach for their business and justify your recommendation.

Provide a detailed comparison of the JIT and EOQ strategies, addressing the following:

- Explain the key principles and assumptions behind the JIT and EOQ approaches.
- Identify the potential advantages and disadvantages of each strategy for the given business scenario.
- Justify which inventory management strategy you would recommend for this business and explain your reasoning.

💡 AI Tip: Ask an AI to explain the underlying principles of JIT and EOQ in a different way. Did this deepen your understanding for the justification you need to provide?

```python
# JIT (Just-In-Time) is an inventory strategy where stock is ordered only when it is needed, aiming to reduce storage costs and waste. It assumes a very reliable supply chain and consistent delivery timing. EOQ (Economic Order Quantity) is a method that calculates the optimal order size to minimise total inventory costs, balancing ordering costs and holding costs. It assumes steady demand and stable costs.

# The main advantage of JIT is low storage cost and reduced waste, but it is risky because delays in supply can cause stock shortages. EOQ provides more stability and ensures stock is always available, but it can lead to higher holding costs and excess inventory. For a small retail business, EOQ is usually more suitable because it is safer and more reliable, especially if demand is not perfectly stable. JIT may only work well if the supplier is very reliable and demand is predictable. Therefore, I would recommend EOQ as it better balances cost control and stock availability.
```

### Extension: Integrating Concepts

Your business has now grown, and the owner wants to expand the inventory management system to include more advanced features, such as automated reordering and stock level forecasting.

Design an enhanced dictionary-powered inventory management system that incorporates these new requirements. In your design, explain:

- How you would modify the dictionary structure to accommodate the additional features.
- What algorithms or techniques you would use for automated reordering and forecasting, and justify your choices.
- Potential challenges you might face when implementing this expanded system and how you would address them.

```python
# To expand the inventory system with automated reordering and forecasting, I would enhance the dictionary structure by adding extra fields such as minimum stock level, reorder point, lead time, and sales history. The structure could look like: {product_id: {"name": ..., "stock": ..., "min_stock": ..., "reorder_point": ..., "sales_history": [...]}}, which allows both real-time tracking and predictive analysis.
# For automated reordering, I would use a simple rule-based algorithm: if stock <= reorder_point, the system automatically triggers a restock order based on EOQ or a fixed order quantity. For forecasting, I would use a basic moving average method on the sales history to predict future demand, as it is simple and effective for small to medium datasets.
# Some challenges include handling large amounts of historical data, increased memory usage from storing sales history in dictionaries, and reduced accuracy in forecasting if demand is irregular. To address this, I would limit stored history (e.g., last 30 days), or use external databases for scalability. Overall, while dictionaries are efficient for fast lookups and updates, combining them with simple forecasting algorithms helps balance performance and functionality.
```
